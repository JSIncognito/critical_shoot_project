#### 3.결론

     * 가설1 : 단계별 거리두기를 시행하였을때, 높은 단계일수록 대중교통 이용량이  줄어들 것이다.(통계자료 토대로 결과 도출)

     * 가설4 : 연령별 대중교통 이용량 변화 
      - ex) 거리두기 단계가 올라갔을때 어떤한 연령층의 이용량이 증가하고 감소하였지 비교.

````python
death = pd.read_excel(path_1 + "기본통계_질병별_사망.xlsx", skiprows=4,thousands = ',')
month = pd.read_excel(path_1 + "기본통계_질병별_월별.xlsx", skiprows=4,thousands = ',')
covid = pd.read_excel(path_1 + "covid19(월별확진자).xlsx")

df_death = death.iloc[:-2 , 2:-1]
df_month = month.iloc[:-2, 2:-1]

df_death.rename(columns={"구분.1" : "질병"}, inplace=True)
df_month.rename(columns={"구분.1" : "질병"}, inplace=True)
````

````python
#%%

covid_month.index = covid_month["Unnamed: 0"]
covid_month.index.name = "년도"
covid_month.drop(["Unnamed: 0", "총합"], axis = 1, inplace=True)
covid_month = covid_month.T

covid_month
````

````python
#%%

# 월별 코로나 확진자 수 리스트 생성

covid_month_2020 = covid_month[2020]
covid_month_2021 = covid_month[2021]

total_covid = pd.concat([covid_month_2020,covid_month_2021])
df = pd.DataFrame(total_covid, columns=["확진자 발생"])

covid_month_list = [] # 코로나 확진자 발생 수를 담을 list

for i in df.values[:-5]:
    covid_month_list.append(i[0])
````

````python
#%%

# 월 인덱스 생성

covid_month_index = []

for i in range(2020,2022):
    if i == 2020:
        for j in range(1,13):
            covid_month_index.append(str(i)+'-'+str(j))
    else:
        for k in range(1,8):
            covid_month_index.append(str(i)+"-"+str(k))
````

````python
#%%

# 월별 코로나 확진자 DataFrame 생성

covid_list = [covid_month_index,covid_month_list]
covid_df = pd.DataFrame(covid_list,index=["Month","Number of Coronic"]).T
covid_df
````

````python
#%%

# Month column datatype => datetime으로 변환
covid_df["Month"] = pd.to_datetime(covid_df["Month"])

# 단계표기를 위해 Step 컬럼 생성
covid_df["Step"] = "" 

# 1단계 표기

# if (covid_df["Month"] > "2020-02") & (covid_df["Month"] < "2020-09"):
#     covid_df["Step"] = "1단계"
a = (covid_df["Month"] > "2020-02") & (covid_df["Month"] < "2020-09")
b = (covid_df["Month"] == "2020-09")
c = (covid_df["Month"] > "2020-09") & (covid_df["Month"] < "2020-12")
d = (covid_df["Month"] > "2020-11") & (covid_df["Month"] < "2021-03")
e = (covid_df["Month"] > "2021-02")& (covid_df["Month"] < "2021-07")

for i in range(len(a)):
    if a[i]:
        covid_df["Step"][i] = "1단계"
     
for i in range(len(b)):
    if b[i]:
        covid_df["Step"][i] = "2단계"
     
for i in range(len(c)):
    if c[i]:
        covid_df["Step"][i] = "1단계"
     
for i in range(len(d)):
    if d[i]:
        covid_df["Step"][i] = "2.5단계"
     
for i in range(len(e)):
    if e[i]:
        covid_df["Step"][i] = "2단계"
````

````python
#%%

# 월별 코로나 확진자 수 그래프

fig = px.line(covid_df,
             x = "Month",
             y = "Number of Coronic",
             title= "월별 코로나 확진자 수"
             )

fig
````

````python
#%%

# 단계에 따른 확진자 수 

fig = px.bar(covid_df, x="Month", y='Number of Coronic',color = "Step")
fig
````

````python
#%%
covid_step = pd.read_csv(path + "covid_step.csv", index_col="Unnamed: 0")
covid_step
````

````python
#%%

# 단계별 증감율 구하기

step_updown = []

a = list(covid_step["평균 확진자 수"].values)
b = list(covid_step["다음달 확진자"].values)

# 증감률
step_updown=[
            round((b[0] - a[0])/b[0] * 100,1),
            round((b[1] - a[1])/b[1] * 100,1),
            round((b[2] - a[2])/b[2] * 100,1),
            round((b[3] - a[3])/b[3] * 100,1),
            round((b[4] - a[4])/b[4] * 100,1)
]
step_updown
````

````python
#%%

covid_step["증감률"] = step_updown
covid_step.index.name="단계"
covid_step.reset_index(inplace=True)
covid_step
````

````python
#%%

px.line(covid_step,x = "단계",y = "증감률",
        title = "단계별 증감율",
        )
````

````python
a = subway_month_sum[(subway_month_sum["사용월"]<202001)]
b = subway_month_sum[(subway_month_sum["사용월"]<202101)& (subway_month_sum["사용월"]>201912)]
c = subway_month_sum[(subway_month_sum["사용월"]>202012)]

# 월을 문자열로 변환
def toStrMonth(a):
    a_list = []
    pat = re.compile("(\d{4})(\d{2})")
    for i in a["사용월"]:
        a_ = pat.sub("\g<2>월",str(i))
        for i in range(len(a_)):
            if a_[0] == "0":
                a_ = a_[1:]
        a_list.append(a_)
    a["사용월"] = a_list
    
toStrMonth(a)
toStrMonth(b)
toStrMonth(c)
````
