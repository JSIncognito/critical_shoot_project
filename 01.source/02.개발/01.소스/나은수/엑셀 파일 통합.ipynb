```python
import pandas as pd
import glob
import re
```

```python
path = 'C:/Users/NaEunSu/project01/응용지표_2019/'
files = glob.glob(path + "*.xlsx")
excel = pd.DataFrame()
```

```python
for file_name in files:
    df = pd.read_excel(file_name)
    print(df["일"])
    excel = pd.concat([excel,df],ignore_index=True)
```

```python
excel = excel.iloc[1::2,:]
excel
```

```python
# excel.to_excel("20200101-20200731.xlsx")
```

```python
excel["일"]
```

```python
day = excel["일"]
day
```

```python
day_list = []
for d in day:
    d = re.sub("\(\w\)","",d)
    day_list.append(d)
```

```python
excel["일"] = day_list
```

```python
excel["일"] = pd.to_datetime(excel["일"])
```

```python
excel["일"]
```

```python
excel.sort_values(by="일", inplace=True)
```

```python
excel = excel.reset_index()
excel
```

```python
del excel["index"]
```
