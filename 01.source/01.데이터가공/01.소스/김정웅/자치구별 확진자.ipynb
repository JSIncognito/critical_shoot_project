```python
df=pd.read_csv('./서울특별시 코로나19 자치구별 확진자 발생동향.csv', encoding='cp949')
```

```python
df3 = []
for i in df.iloc[0].index:
    if '전체' in i:
        df3.append(pd.DataFrame({'자치구' : i,
                     '확진자' : df.iloc[0][i]}, index = [0]))
    else:
        pass
    
```

```python
df4=pd.concat(df3)
```

```python
df4.to_excel('자치구 전체 확진자.xlsx')
```

```python
for i in range(len(df)):
    df.loc[i, '자치구 기준일'] = df.loc[i, '자치구 기준일'][:7]
```

```python
df_day=list(df['자치구 기준일'].unique())
```

```python
df5 = []
df6 = []
for i in range(len(df_day)):
    df[df['자치구 기준일'] == df_day[i]]
    for j in df.iloc[0].index:
        if '전체' in j:
            df5.append({'월' : df_day[i],
                        '자치구' : j,
                        '확진자' : df.iloc[0][j]})
   
    df6.append(df5)
```

```python
df7 = []
df9 = []
for i in range(len(df_day)):
    df12=df[df['자치구 기준일'] == df_day[i]]
    for j in df.columns[1:]:
        if '추가' in j:
            df7.append({'월' : df_day[i],
                        '자치구' : j,
                        '확진자' :str(df12[j].sum())})
    df9.append(df7)
```

```python
df10=pd.DataFrame(df9[0])
```
