info 확인 결과 대여건수가 대여건수 컬럼에 464개, Unnamed: 2 컬럼에 416개로 나눠져 있었음

그래서 대여건수와 Unnamed: 2를 합쳐주는 과정을 진행하겠음

```python
df["대여건수"] = df["대여건수"].fillna(0)
```

```python
df["Unnamed: 2"] = df["Unnamed: 2"].fillna(0)
```

```python
df["대여건수"] = df["대여건수"]+df["Unnamed: 2"]
```

대여건수에 880개로 합쳐져있는걸 확인

```python
# 불피요한 컬럼 제거
df = df.iloc[ : , :2]
df
```

```python
# 대여일자를 행 인덱스로 설정
df.set_index(df['대여일자'], inplace=True)
```

```python
df
```

```python
# month = df["대여건수"].resample("M").sum()
```

```python
# 대여건수만 Dataframe으로 가져가기
month = df["대여건수"]
month
```

```python
month = month.to_frame()
```

```python
# 행 인덱스에서 뒤에 일이 나와있는데 이것을 제거하기위한 코드
new_mon = []
for mon in month.index:
    mon = str(mon)
    new_mon.append(re.search("[\d]+-[\d]+",mon).group())
```

```python
new_mon
```

```python
month.index
```

```python
month.index = new_mon
month
```
